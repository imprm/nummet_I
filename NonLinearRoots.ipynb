\begin{equation}
2 Q - ET = 0
\end{equation}

\begin{equation}
Q = c (h_L - h)
\end{equation}

\begin{equation}
2 c (h_L - h) - ET(h) = 0
\end{equation}

\begin{equation}
h = \frac{2 c h_L - ET(h)}{2 c} = f_p(h)
\end{equation}

and is nonlinear due to the dependence on

\begin{equation}
ET(h) = et_{max} erf \left( \frac{4h}{h_L} \right)
\end{equation}

```python
c = 1.
h_L = 10.
et_max = 15.
h_0 = h_L/2.
```

```python
def ET(h):
    return et_max*scis.erf(4*h/h_L)
```

```python
h = np.linspace(0, h_L, 50)
et = ET(h)
fig, ax = plt.subplots()
ax.plot(h, et, color='r')
ax.set_ylim(0,2*et_max)
ax.set_xlabel('$h$')
ax.set_ylabel('$ET(h)$')
plt.show()
```

```python
def fp(h):
    return (2*c*h_L - ET(h))/(2*c)
```

```python
h = h_0
tol = h_L/50000.
fp_p = []
h_p = []
while 1:
    print 'Iteration {0:2d} yields h = {1:7.5f}'.format(len(h_p), h)
    h_p.append(h)
    h = fp(h)
    fp_p.append(h)
    if abs(h_p[-1]-fp_p[-1]) < tol:
        break
print 'Iteration {0:2d} yields h = {1:7.5f}'.format(len(h_p), h)
```

```python
esenummet.plot_picard_convergence_pattern(h_p, fp_p)
```

\begin{equation}
h = \frac{2 c h_L - ET(h)}{2 c}
\end{equation}

\begin{equation}
0 = \frac{2 c h_L - ET(h)}{2 c} - h = f_n(h)
\end{equation}
\begin{equation}
f_n(h) = f_p(h) - h
\end{equation}

```python
def fn(h):
    return fp(h) - h
```

```python
h = np.linspace(0., h_L, 100)
f = fn(h)
```

```python
fig, ax = plt.subplots()
ax.plot(h, f, color='r')
ax.set_xlabel('$h$')
ax.set_ylabel('$f_n(h)$')
plt.show()
```

\begin{equation}
f_n'(h) \approx \frac{f_n(h+dh) - f_n(h)}{dh}
\end{equation}

```python
dh = h_L/100.
h = h_0
h_n = [h]
while 1:
    dfdh = (fn(h+dh) - fn(h)) / dh
    a = fn(h)-(dfdh*h)
    h = -a/dfdh
    h_n.append(h)
    print 'Iteration {0:1d} at h = {1:7.5f} yields dfdh = {2:5.3f} and new h = {3:7.5f}'.format(len(h_n)-1, h_n[-2], dfdh, h)
    if abs(h_n[-1]-h_n[-2]) < tol:
        break
```

So the Newton method converges in 5 iterations compared to 30 for the Picard method. But what happened inside a single iteration? For the Newton method, we evaluate $f_n(h)$ two times (at $h$ and $h+dh$), and for a Picard iteration just once. The quadratic convergence of the Newton method comes at a cost, but still comes out ahead in terms of efficiency.

```python
esenummet.plot_newton_convergence_pattern(fn, dh, h_0, tol)
```

```python
print scop.newton(fn, h_0)
```

```python
def rootsearch(f, a, b, dx):
    """
    Searches the interval (a,b) in increments dx for
    the bounds (x1,x2) of the smallest root of f(x).
    """
    x1 = a
    f1 = f(a)
    x2 = a + dx
    f2 = f(x2)
    print 'f({0:3.1f}) * f({1:3.1f}) = {2:5.1f}'.format(x1,x2,f1*f2)
    while True:
        x1 = x2
        f1 = f2
        x2 = x1 + dx
        f2 = f(x2)
        print 'f({0:3.1f}) * f({1:3.1f}) = {2:5.1f}'.format(x1,x2,f1*f2)
        if f1*f2 < 0.:
            break
    return x1,x2
```

```python
def bisection(f, x1, x2, tol=1.0e-5):
    """
    Finds a root of f(x) = 0 by bisection.
    The root must be bracketed in (x1,x2).
    """
    f1 = f(x1)
    f2 = f(x2)
    while abs(x1-x2) > tol:
        x3 = 0.5*(x1 + x2)
        f3 = f(x3)
        print 'x = {0:8.6f}, f(x) = {1:13.10f} between [{2:8.6f},{3:8.6f}]'.format(x3,f3,x1,x2)
        if f2*f3 < 0.0:
            x1 = x3
            f1 = f3
        else:
            x2 = x3
            f2 = f3
    return (x1 + x2)/2.0
```

Now that we know the root is bracketed in $[a,b]$ we can use a bisection algorithm

```python
root = bisection(foo, a, b)
print root
```

```python
def ridder(f, a, b, tol=1.0e-9, max_iter=30):
    """
    Finds a root of f(x) = 0 with Ridder’s method to a tolerance of tol.
    The root must be bracketed in (a,b).
    """
    fa = f(a)
    fb = f(b)
    for i in range(max_iter):
        # Compute the improved root x from Ridder’s formula
        c = 0.5*(a + b)
        fc = f(c)
        s = np.sqrt(fc**2 - fa*fb)
        dx = (c - a)*fc/s
        if (fa - fb) < 0.0:
            dx = -dx
        x = c + dx
        fx = f(x)
        print 'Iteration {0:1d} at x = {1:7.5f} yields fx = {2:7.3f}'.format(i, x, fx)
        # Test for convergence
        if i > 0:
            if abs(x - xOld) < tol*max(abs(x),1.0):
                return x
        xOld = x
        # Re-bracket the root as tightly as possible
        if fc*fx > 0.0:
            if fa*fx < 0.0:
                b = x
                fb = fx
            else:
                a = x
                fa = fx
        else:
            a = c
            b = x
            fa = fc
            fb = fx
```

```python
print 'Bracketing the root:'
a, b = rootsearch(foo, 0., 1., 0.2)
print a, b

print '\nFinding the root using ridders method:'
root = ridder(foo, a, b)
print root
```

```python
def newton_raphson(f, x0, tol=1.48E-08, dx=1.0E-9):
    x = x0
    x_n = [x]
    while 1:
        dfdx = (f(x+dx) - f(x)) / dx
        a = f(x)-(dfdx*x)
        x = -a/dfdx
        x_n.append(x)
        print 'Iteration {0:1d} at x = {1:7.5f} yields dfdx = {2:7.3f} and new x = {3:7.5f}'.format(len(x_n)-1, x_n[-2], dfdx, x)
        if abs(x_n[-1]-x_n[-2]) < tol:
            break
    return x
```
